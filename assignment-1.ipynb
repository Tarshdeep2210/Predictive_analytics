```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "/kaggle/input/india-air-quality-data/data.csv",
    encoding="latin1"
)

x = df["no2"].dropna().values

print("Total NO2 samples:", len(x))
r = 102316050
a = 0.05 * (r % 7)
b = 0.3 * (r % 5 + 1)

print("a_r =", a)
print("b_r =", b)

z = x + a * np.sin(b * x)

mu = np.mean(z)

sigma1 = np.var(z)

lam = 1 / (2 * sigma1)
c = 1 / np.sqrt(2 * np.pi * sigma1)
print("\nFINAL PARAMETERS ")
print("μ (mu) =", mu)
print("λ (lambda) =", lam)
print("c   =", c)
plt.figure(figsize=(8,5))
plt.hist(z, bins=50, density=True)
plt.title("PDF of Transformed Variable z")
plt.xlabel("z")
plt.ylabel("Probability Density")
plt.show()
```
